# Manifest generation

This notebook prepares the manifest file for the pipeline

The format of the manifest is like:
```
- summaryStatisticsPath
- trait
- sampleSize
- ancestry
```

````python
from pathlib import Path
from typing import TypedDict
import pandas as pd
````

````python
eur = Path("testdata/eur_gwas/studyId=5z4VeOjnJy/*.parquet")
afr = Path("testdata/afr_gwas/studyId=Jgq8sXKNio/*.parquet")
````

````python
class ManifestRow(TypedDict):
    summaryStatisticsPath: str
    trait: str
    sampleSize: int
    ancestry: str
````

````python
manifest = pd.DataFrame.from_records(
    [
        # 2 traits -> 2 MA runs
        ManifestRow(summaryStatisticsPath=eur, trait="A", sampleSize=30_000, ancestry="EUR"),
        ManifestRow(summaryStatisticsPath=afr, trait="A", sampleSize=30_000, ancestry="AFR"),
        ManifestRow(summaryStatisticsPath=eur, trait="B", sampleSize=30_000, ancestry="EUR"),
        ManifestRow(summaryStatisticsPath=afr, trait="B", sampleSize=30_000, ancestry="AFR"),
    ]
)
````

````python
manifest
````

````python
manifest.to_csv("../../testdata/manifest.tsv", sep="\t", index=False)
````
