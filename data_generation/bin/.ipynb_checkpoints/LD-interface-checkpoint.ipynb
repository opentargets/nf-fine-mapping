# LD manifest

This notebook creates LD index manifest

```python
import pandas as pd
from enum import StrEnum
```

```python
output_ld_manifest_path = "../../testdata/ld_ancestry.tsv"
```

```python
ld_matrix_template = "gs://gcp-public-data--gnomad/release/2.1.1/ld/gnomad.genomes.r2.1.1.{POP}.common.adj.ld.bm"
ld_index_raw_template = (
    "gs://gcp-public-data--gnomad/release/2.1.1/ld/gnomad.genomes.r2.1.1.{POP}.common.ld.variant_indices.ht"
)
```

```python
class LDAncestry(StrEnum):
    AFR = "AFR"
    AMR = "AMR"
    EAS = "EAS"
    FIN = "FIN"
    NFE = "NFE"
```

```python
data = pd.DataFrame(
    [
        {
            "ancestry": ancestry.value,
            "ldMatrix": ld_matrix_template.format(POP=ancestry.value),
            "ldIndex": ld_index_raw_template.format(POP=ancestry.value),
        }
        for ancestry in LDAncestry
    ]
)
```

```python
data.to_csv("../../testdata/ld_reference.tsv", sep="\t", index=False)
```
