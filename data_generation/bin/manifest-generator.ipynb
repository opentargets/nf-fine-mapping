# Manifest generation

This notebook prepares the manifest file for the pipeline

The format of the manifest is like:
```
- summaryStatisticsPath
- trait
- sampleSize
- ancestry
```

````python
from pathlib import Path
from typing import TypedDict
import pandas as pd
````

````python
eur = Path("../data/eur_gwas/studyId=n0e179d3td/*.parquet")
afr = Path("../data/afr_gwas/studyId=ye8T5hqmrR/*.parquet")
````

````python
class ManifestRow(TypedDict):
    summaryStatisticsPath: str
    trait: str
    sampleSize: int
    ancestry: str
    
    
````

````python
manifest = pd.DataFrame.from_records([
    ManifestRow(summaryStatisticsPath=eur, trait="A", sampleSize = 30_000, ancestry="EUR"),
    ManifestRow(summaryStatisticsPath=afr, trait="A", sampleSize = 30_000, ancestry="AFR"),
])
````

````python
manifest
````

````python
manifest.to_csv("../data/manifest.tsv", sep='\t')
````
